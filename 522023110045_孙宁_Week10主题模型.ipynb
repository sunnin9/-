## è¯­æ–™é¢„å¤„ç†

```python
import pandas as pd #æ•°æ®è¡¨
import numpy as np #æ•°ç»„
import re #æ­£åˆ™è¡¨è¾¾å››
import jieba #ä¸­æ–‡åˆ†è¯
import matplotlib.pyplot as plt #ç”»å›¾
from gensim import corpora, models 
import pyLDAvis #äº¤äº’å¼LDAå¯è§†åŒ–
import pyLDAvis.gensim_models as gensimvis
```

```python
df = pd.read_csv('text_analysis_weibo.csv', index_col = 0)
```

```python
df.head()
```

### å‰”é™¤ç¬¦å·ä¸æ•°å­—

```python
def remove_nums(text):
    nonums = re.sub('[^\u4e00-\u9fa5]+', '', text)
    return nonums
test = df['æ ‡é¢˜/å¾®åšå†…å®¹'][0]
remove_nums(test)
```

### åˆ†è¯

```python
def clean_text(text):
    words = jieba.lcut(text)
    return ' '.join(words)

test = df['æ ‡é¢˜/å¾®åšå†…å®¹'][0]
clean_text(test)
```

```python
df['æ ‡é¢˜/å¾®åšå†…å®¹'] = df['æ ‡é¢˜/å¾®åšå†…å®¹'].astype(str)
df['å¾®åšå†…å®¹åˆ†è¯'] = df['æ ‡é¢˜/å¾®åšå†…å®¹'].apply(remove_nums)
df['å¾®åšå†…å®¹åˆ†è¯'] = df['å¾®åšå†…å®¹åˆ†è¯'].apply(clean_text)
df['å¾®åšå†…å®¹åˆ†è¯'] = df['å¾®åšå†…å®¹åˆ†è¯'].apply(lambda x: x.split())
df[:20]
```

## LDA

```python
dictionary = corpora.Dictionary(df['å¾®åšå†…å®¹åˆ†è¯']) #æ ¹æ®åˆ†è¯ç»“æœåˆ›å»ºå­—å…¸
corpus = [dictionary.doc2bow(text) for text in df['å¾®åšå†…å®¹åˆ†è¯']] #æ ¹æ®åˆ†è¯ç»“æœåˆ›å»ºè¯­æ–™åº“
```

```python
# è®­ç»ƒLDAæ¨¡å‹
lda_model = models.LdaModel(corpus, num_topics=8, id2word=dictionary, passes=15)
```

```python
# æŸ¥çœ‹ä¸»é¢˜
topics = lda_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
```

## å¯è§†åŒ–

```python
lda_vis = gensimvis.prepare(lda_model, corpus, dictionary, n_jobs=1) 

#æ˜¾ç¤ºå¯è§†åŒ–ç•Œé¢
pyLDAvis.display(lda_vis)
```

```python
#å¯¼å‡ºå¯è§†åŒ–ç»“æœåˆ°html
pyLDAvis.save_html(lda_vis, 'lda_visualization.html')
```
