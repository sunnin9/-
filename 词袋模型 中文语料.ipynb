```python
import pandas as pd
```

```python
#导入数据
text_ana = pd.read_csv('text_analysis_weibo.csv')
text_ana = text_ana.drop(['Unnamed: 0'],axis=1)
text_ana
```

```python
#分析语料
text = list(text_ana['标题/微博内容'])
text
```

```python
#jieba分词
```

```python
#pip install jieba
import jieba
```

```python
def word_segmentation(text):
    words = jieba.lcut(text)
    return ''.join(words)

text_ana['标题/微博内容'] = text_ana['标题/微博内容'].iloc[:100]\
.astype(str).apply(word_segmentation)
text_ana['标题/微博内容'].head()
```

```python
text = list(text_ana['标题/微博内容'].dropna())
text
```

```python
#构建词袋模型
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#初始化CountVectorizer
vectorizer = CountVectorizer()
```

```python
#将文本数据转换为词袋特征矩阵
X = vectorizer.fit_transform(text)
```

```python
#展示特征矩阵的形状
print('词袋特征矩阵的型庄:', X.shape)
```

```python
#展示词袋模型中的所有特征词
print('特征词列表:', vectorizer.get_feature_names_out())
```

```python
list(X.toarray())
```

```python
#词频统计
```

```python
#获取词汇表
feature_words = vectorizer.get_feature_names_out()

#词频统计
word_freq = dict(zip(feature_words, X.sum(axis=0).A1))
```

```python
#观察词袋模型的词频计数结果
sorted(word_freq.items(), key=lambda x:x[1], reverse=True)
```

```python
#绘制词云图
```

```python
from wordcloud import WordCloud
```

```python
import matplotlib.pyplot as plt
```

```python
# 根据词袋模型分析结果绘制词云图
wc = WordCloud(width=1200,
               height=800,
               background_color='white',
               font_path='simhei.ttf').generate_from_frequencies(word_freq)
```

```python
# 显示词云图
plt.figure(figsize=(10,5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
